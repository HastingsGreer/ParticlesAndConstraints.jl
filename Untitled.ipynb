```python
def ind(n):
    return np.index_exp[2 * n:2 * n + 2]

class Slider():
    def __init__(self, particle, normal):
        normal = torch.tensor(normal).float()
        normal = normal / torch.linalg.norm(normal)
        self.particle = particle
        self.normal = normal
        self.oneway=False
    def compute_effect(self, system, column):
        column[ind(self.particle)] = self.normal
    def compute_acceleration(self, system):
        return torch.sum(system.forces[ind(self.particle)] * self.normal)
    def compute_conserved_quantity(self, system):
        return torch.sum(system.positions[ind(self.particle)] * self.normal)
class Shelf(Slider):
    def __init__(self, particle, normal):
        normal = torch.tensor(normal).float()
        normal = normal / torch.linalg.norm(normal)
        self.particle = particle
        self.normal = normal
        self.oneway=True
class Rod():
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        self.oneway=False
    def compute_effect(self, system, column):
        direction = system.positions[ind(self.p1)] - system.positions[ind(self.p2)]
        direction = direction / torch.linalg.norm(direction)  
        column[ind(self.p1)] = direction
        column[ind(self.p2)] = -direction
    def compute_acceleration(self, system):
        displacement = system.positions[ind(self.p1)] - system.positions[ind(self.p2)]
        length = torch.linalg.norm(displacement)  
        v = system.velocities[ind(self.p1)] - system.velocities[ind(self.p2)]    
        return -torch.sum(v * v) / length
    def compute_conserved_quantity(self, system):
        displacement = system.positions[ind(self.p1)] - system.positions[ind(self.p2)]
        length = torch.linalg.norm(displacement) 
        return length
class Prop(Rod):
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        self.oneway=True
class System():
    def __init__(self, n_particles, constraints, masses):
        self.n_particles = n_particles
        self.forces = torch.tensor(list(zip(np.zeros(self.n_particles), np.ones(self.n_particles)))).flatten().float()
        self.constraints = constraints
        
        self.masses = (torch.tensor(masses)[:, None] * torch.ones((1, 2))).flatten()
        
    def dvdt(self):
        interactions = torch.zeros(len(self.constraints), 2 * self.n_particles)
        for i, constraint in enumerate(self.constraints):
            constraint.compute_effect(self, interactions[i])     
        interactions2 = interactions / self.masses[None, :] @ interactions.transpose(0, 1)       
        desires = torch.cat([constraint.compute_acceleration(self).reshape(1) for constraint in self.constraints])
        constraint_forces = torch.linalg.solve(interactions2[None], desires[None])
        for i, (c, f) in enumerate(zip(self.constraints, constraint_forces[0])):
            if f.item() < 0 and c.oneway:
                del self.constraints[i]
        dvdt = constraint_forces @ interactions / self.masses - self.forces
    
        return dvdt[0]
    
    def dydt(self, y):
        self.positions = y[:2 * self.n_particles]
        self.velocities = y[2 * self.n_particles:]
        return torch.cat([self.velocities, self.dvdt()])
    
    def compute_conserved_quantities(self, y):
        self.positions = y[:2 * self.n_particles]
        self.velocities = y[2 * self.n_particles:]
        return [c.compute_conserved_quantity(self) for c in self.constraints]
    
    def rk4(self, y_0, t_final, h=.01):
        
        y = y_0

        t = 0
        
        output = [y[None]]
        
        constraints_copy = self.constraints[:]
        
        while t < t_final and y[8] < 3.7:
            t += h
            
            k1 = self.dydt(y)
            k2 = self.dydt(y + h * k1 / 2)
            k3 = self.dydt(y + h * k2 / 2)
            k4 = self.dydt(y + h * k3)

            y = y + h * 1/6 * (k1 + 2 * k2 + 2 * k3 + k4)
            
            output.append(y[None])
            
        self.constraints = constraints_copy
        return torch.cat(output)
    
    def rk40(self, r_0, t_final, h=0.01):
        if isinstance(r_0, list):
            r_0 = torch.tensor(r_0).float()
        y_0 = torch.cat([r_0, 0 * r_0])
        return self.rk4(y_0, t_final, h=h)
    
    def energy(self, y):
        return torch.sum(self.masses * (self.forces * y[:2 * self.n_particles] + 
                                        .5 * y[2 * self.n_particles:]**2))
    
    def render(self, solution):
        spacing = 10
        plt.plot(solution[::spacing, list(range(0, self.n_particles * 2, 2))].transpose(0, 1), 
                 solution[::spacing, list(range(1, 1 + self.n_particles * 2, 2))].transpose(0, 1))
        plt.gca().set_aspect("equal")
        plt.show()
      
```
